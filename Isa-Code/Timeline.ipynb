```python
#First step is to organize this data by the date-time variable and location 
#check variable types, we need to convert time into a datetime variable
reports.dtypes
```

```python
#converting Timestamp column from object to datetime 
reports['time'] = pd.to_datetime(reports['time'], format='%Y-%m-%d %H:%M:%S')
```

```python
#checking if conversion was succesful -- it was 
reports.dtypes
```

```python
reports
```

```python
#organize by dates , looks like we have datesfrom April 6 to April 11
reports = reports.sort_values(by='time',ascending=True)
```

```python
reports.groupby(['time'], as_index = False).mean()
```

```python
reports['location'].unique()
```

```python
#Filter by location 1
reports_1 = reports[reports['location'] == 1]
reports_1.groupby(['time'], as_index = False).mean()
```

```python
#ok so idk how to get time in yet but I want to look at the magnitude of earthquake on each day for now 
reports = reports.groupby(['location', 'time'], as_index=False).mean()
```

```python
reports.to_csv("C:\\Users\\igome\\Documents\\Smith\\Spring_2021\\visual analyticsd\\mc1-data-reports-grouped.csv")
```

```python
# pivoting the dataframe
reports_melted = reports.melt(id_vars=["location", "time"], 
        var_name="category", 
        value_name="intensity")
```
